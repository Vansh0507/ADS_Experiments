```python
import altair as alt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, Range1d
```

```python
# Group by 'race/ethnicity' and find the maximum math score for each group
max_math_scores = df.groupby('race/ethnicity')['math score'].max()

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(max_math_scores.index, max_math_scores.values, marker='o', linestyle='-')

# Adding titles and labels
plt.title('Max Math Score by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Maximum Math Score')
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python

# Example: Subplot with 2 rows and 2 columns
plt.figure(figsize=(10, 10))
# Plot 1: Distribution of Math Score
plt.subplot2grid((2, 2), (0, 0))
plt.hist(df['math score'], bins=20, color='skyblue')
plt.title('Distribution of Math Score')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

# Plot 2: Distribution of Reading Score
plt.subplot2grid((2, 2), (0, 1))
plt.hist(df['reading score'], bins=20, color='lightgreen')
plt.title('Distribution of Reading Score')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

# Plot 3: Distribution of Writing Score
plt.subplot2grid((2, 2), (1, 0))
plt.hist(df['writing score'], bins=20, color='salmon')
plt.title('Distribution of Writing Score')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')

# Plot 4: Distribution of Gender
plt.subplot2grid((2, 2), (1, 1))
df['gender'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Create subplots
plt.figure(figsize=(18, 6))

# Plot 1: Distribution of Math Scores
plt.subplot(1, 3, 1)
sns.histplot(data=df, x='math score', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.grid(True)

# Plot 2: Distribution of Reading Scores
plt.subplot(1, 3, 2)
sns.histplot(data=df, x='reading score', bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.grid(True)

# Plot 3: Distribution of Writing Scores
plt.subplot(1, 3, 3)
sns.histplot(data=df, x='writing score', bins=30, kde=True, color='salmon')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Get the counts of students for each race/ethnicity
race_count = df['race/ethnicity'].value_counts()

# Define colors
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  # You can add more colors if needed

plt.figure(figsize=(8, 8))
plt.pie(race_count, labels=race_count.index, colors=colors,
        startangle=90, autopct='%1.1f%%')
plt.title('Distribution of Students by Race/Ethnicity')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.kdeplot(data=df, x='math score', y='reading score', fill=True, cmap='Blues', shade=True)
plt.title('2D Density Plot of Math Score vs. Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, Range1d

# Calculate the average math score by race/ethnicity
avg_math_by_race = df.groupby('race/ethnicity')['math score'].mean().reset_index()

# Create DataFrame for count of students by race/ethnicity
race_counts = df['race/ethnicity'].value_counts().reset_index()
race_counts.columns = ['race/ethnicity', 'Count']

# Concatenate average math score and count DataFrames
data = pd.concat([avg_math_by_race, race_counts['Count']], axis=1)

# Remove the index column from the DataFrame
data = data[['race/ethnicity', 'math score', 'Count']]

# Create Bokeh figure
p = figure(x_range=data['race/ethnicity'].tolist(), y_range=Range1d(0, max(data['Count']) * 1.1),
           width=800, height=400,
           toolbar_location=None, tools='hover',
           tooltips=[('Race/Ethnicity', '@race/ethnicity'), ('Average Math Score', '@{math score}{0.2f}'), ('Count', '@Count')])

# Scale the circle sizes based on the average math score
max_score = data['math score'].max()
min_score = data['math score'].min()
size_scale = 30  # Adjust the scaling factor as needed
min_circle_size = 5  # Minimum size for circles
data['circle_size'] = size_scale * ((data['math score'] - min_score) / (max_score - min_score)) + min_circle_size

# Plot circles for each race/ethnicity with size representing the average math score
source = ColumnDataSource(data)
p.circle(x='race/ethnicity', y='Count', size='circle_size', source=source, fill_alpha=0.6, line_color='black', line_width=2)

# Customize plot
p.xaxis.major_label_orientation = 1.2  # Rotate x-axis labels
p.xaxis.major_label_text_align = 'right'  # Align x-axis labels to the right
p.yaxis.axis_label = 'Count of Students'
p.xaxis.axis_label = 'Race/Ethnicity'

# Show plot in notebook
output_notebook()
show(p)
```
